```python
import pandas as pd
import csv
```

```python
with open('/content/Raj_coordinates.csv', encoding='utf-8-sig') as file_obj:
    reader_obj = csv.DictReader(file_obj)
    for row in reader_obj:
        print(row)
        data_path=row['Loc']
        df = pd.read_csv("IMD_rain_avg/"+data_path+'.csv')
        df2 = pd.read_csv("IMD_tmax_avg/"+data_path+'.csv')
        df3 = pd.read_csv("IMD_tmin_avg/"+data_path+'.csv')
        df['tmax']=df2['Temp']
        df['tmin']=df3['Temp']
        print(df)
        df.to_csv('IMD_combine/'+data_path+'.csv')
```
