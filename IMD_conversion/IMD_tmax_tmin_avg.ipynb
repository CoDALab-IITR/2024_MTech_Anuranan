```python
import pandas as pd
import csv
with open('/content/dailymonthly.csv', encoding='utf-8-sig') as file_obj:

    # Create reader object by passing the file
    # object to DictReader method
    reader_obj = csv.DictReader(file_obj)

    # Iterate over each row in the csv file
    # using reader object
    for row in reader_obj:
        print(row)
        daily=row['daily']
        monthly=row['monthly']
        df = pd.read_csv(daily+'.csv')
        print(df.columns)
        df.columns = ['Date', 'Temp']
        df=df.iloc[:,0:2]
        value = df['Temp'].isnull().sum()
        df['Temp'] = df['Temp'].fillna(0)
        value = df['Temp'].isnull().sum()
        df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
        totalTemp = df.groupby([ df['Date'].dt.year, df['Date'].dt.month ] ).mean({'Temp':sum})
        #temp = df.groupby([ (df['Date']).dt.month ] ).avg({'Temp':sum})
        totalTemp.describe()
        totalTemp.index
        totalTemp.index.names = ['Year', 'Month']
        totalTemp.to_csv(monthly+'.csv')
```
