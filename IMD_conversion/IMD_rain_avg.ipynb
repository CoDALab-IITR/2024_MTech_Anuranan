```python
import pandas as pd
import csv
```

```python
with open('/content/dailymonthly.csv', encoding='utf-8-sig') as file_obj:

    # Create reader object by passing the file
    # object to DictReader method
    reader_obj = csv.DictReader(file_obj)

    # Iterate over each row in the csv file
    # using reader object
    for row in reader_obj:
        print(row)
        daily=row['daily']
        monthly=row['monthly']
        df = pd.read_csv(daily+'.csv')
        print(df.columns)
        df.columns = ['Date', 'PCRP']
        value = df['PCRP'].isnull().sum()
        df['PCRP'] = df['PCRP'].fillna(0)
        value = df['PCRP'].isnull().sum()
        df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
        totalPCRP = df.groupby([ df['Date'].dt.year, df['Date'].dt.month ] ).agg({'PCRP':sum})
        #PCRP = df.groupby([ (df['Date']).dt.month ] ).agg({'PCRP':sum})
        totalPCRP.describe()
        totalPCRP.index
        totalPCRP.index.names = ['Year', 'Month']
        totalPCRP.to_csv(monthly+'.csv')
```
