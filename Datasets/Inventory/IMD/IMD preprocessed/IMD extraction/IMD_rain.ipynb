```python
import imdlib as imd

start_yr = 1960
end_yr = 2023
variable = 'rain' # other options are ('tmin'/ 'tmax')
file_dir = (r'/content/rain') #Path to save the files
data = imd.open_data(variable, start_yr, end_yr,'yearwise', file_dir)
data
```

```python
ds = data.get_xarray()
print(ds)
```

```python
ds = ds.where(ds['rain'] != -999.) #Remove NaN values
ds['rain'].mean('time').plot()
```

```python
with open('/content/Raj_coordinates.csv', encoding='utf-8-sig') as file_obj:

    # Create reader object by passing the file
    # object to DictReader method
    reader_obj = csv.DictReader(file_obj)

    # Iterate over each row in the csv file
    # using reader object
    for row in reader_obj:
        print(row)
        loc=row['Loc']
        x=float(row['x'])
        y=float(row['y'])
        data.to_csv(loc+".csv", x, y, file_dir)
```
